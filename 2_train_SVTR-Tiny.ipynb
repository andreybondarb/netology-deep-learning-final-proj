```python
import matplotlib.pyplot as plt
import regex as re
import numpy as np

file_ = open('output/svtr_rosim_tiny_2_120k/train.log', 'r')
lines = file_.read().splitlines()
file_.close()


losses_ = []
acc_ = []
norm_edit_dis_ = []
for line in lines:

    if not line:
        continue

    loss = re.findall(r'loss: (\d+\.\d+)', line)
    acc = re.findall(r'cur metric, acc: (\d+\.\d+)', line)
    
    if len(loss) != 0:
      losses_.append(float(loss[0]))
    if len(acc) != 0:
      acc_.append(float(acc[0]))
      norm_edit_dis = re.findall(r'norm_edit_dis: (\d+\.\d+)', line)
      norm_edit_dis_.append(float(norm_edit_dis[0]))      
      
print(f'Model name svtr_rosim_tiny 1x_1x_(4498)_(60k_60k_diff_font_sizes)_247_fonts.zip lr 0.0005 warmup_epoch 1 ep 35 bs 256 ch_n 40 mean, std added acc: {round(max(acc_),3)}  mean acc {round(np.mean(acc_),3)} norm_edit_dis_ {round(np.mean(norm_edit_dis_[1:]),3)} {round(max(norm_edit_dis_[1:]),3)}')

plt.plot(losses_[130:])
plt.show()
plt.plot(acc_[130:])
plt.show()
plt.plot(norm_edit_dis_[130:])
plt.show()
```
